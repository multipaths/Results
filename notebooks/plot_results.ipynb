```python
with open('../data/results/pathme_dataset_1.json', 'r') as outfile:
    results = json.load(outfile)
```

```python
aurocs = pd.DataFrame.from_dict(results["auroc_metrics"])
auprcs = pd.DataFrame.from_dict(results["auprc_metrics"])
```

```python
aurocs.rename(columns={
    "raw": "Raw",
    "z": "Z",
    "random_baseline": "Random",
    "page_rank_baseline": "Page-rank",
}, inplace=True)  # more styled names

auprcs.rename(columns={
    "raw": "Raw",
    "z": "Z",
    "random_baseline": "Random",
    "page_rank_baseline": "Page-rank",
}, inplace=True)  # more styled names
```

```python
aurocs.head()
```

```python
sns.boxplot(
    data=aurocs,
)
plt.title("Distribution of AUROCs for each method in Dataset 1")
plt.tight_layout()
plt.savefig(
    os.path.join(os.path.expanduser('~'), 'Desktop', 'aucs_pathme.png'),
    dpi=400,
    bbox_inches='tight', 
)
plt.show()
```

```python
sns.boxplot(
    data=auprcs,
)
plt.title("Distribution of AUPRCS for each method in Dataset 1")
plt.tight_layout()
plt.savefig(
    os.path.join(os.path.expanduser('~'), 'Desktop', 'auprcs_pathme.png'),
    dpi=400,
    bbox_inches='tight', 
)
plt.show()
```
