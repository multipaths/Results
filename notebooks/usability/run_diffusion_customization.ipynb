# Multipaths Usage and Customization

```python
import os
import numpy as np
import pandas as pd

from diffupath.constants import ROOT_RESULTS_DIR, DEFAULT_DIFFUPATH_DIR
```

## DiffuPy

```python
GRAPH_PATH = os.path.join(ROOT_RESULTS_DIR, 'data', 'graphs', '_graph.gml')
DATA_INPUT_PATH = os.path.join(ROOT_RESULTS_DIR, 'data', 'input_samples', 'input_sample_3.csv')
```

#### Input Dataset overview

```python
from diffupy.process_input import process_input_data
```

```python
input_labels = process_input_data(DATA_INPUT_PATH)
pd.DataFrame(index = input_labels.keys(), data={'scores':list(input_labels.values())})
```

#### Graph overview

```python
from diffupy.process_network import process_graph_from_file

import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.read_gml(GRAPH_PATH, label='name')
fig, ax = plt.subplots(1, 1, figsize=(8, 6));
nx.draw_networkx(g, ax=ax)
```

### Run basic diffusion

```python
from diffupy.diffuse import run_diffusion
```

```python
run_diffusion(DATA_INPUT_PATH, GRAPH_PATH).as_pd_dataframe() #DATA INPUT and GRAPH as PATHs
```

```python
run_diffusion(input_labels, g).as_pd_dataframe() #DATA INPUT and GRAPH as Python OBJECTS
```

### Customize diffusion method

#### Available methods

```python
run_diffusion(DATA_INPUT_PATH, GRAPH_PATH, method = 'z').as_pd_dataframe()
```

```python
run_diffusion(DATA_INPUT_PATH, GRAPH_PATH, method = 'raw').as_pd_dataframe()
```

#### Custom method

```python
import networkx as nx
from diffupy.diffuse import run_diffusion

run_diffusion(DATA_INPUT_PATH, GRAPH_PATH, method = nx.pagerank)
```

```python
from diffupy.process_network import process_graph_from_file

GRAPH_PATH = os.path.join(DEFAULT_DIFFUPATH_DIR, 'pickles', 'universe',
                          'pathme_universe_non_flatten_collapsed_names_no_isolates_16_03_2020.pickle')
g = process_graph_from_file(GRAPH_PATH)
```

### Customize kernel calculation

```python
from diffupy.kernels import p_step_kernel
```

```python
run_diffusion(DATA_INPUT_PATH, GRAPH_PATH, kernel_method = p_step_kernel).as_pd_dataframe()
```
